```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import os
import numpy as np
from pathlib import Path


BASE_DIR = Path("DataPaper")
```

1) Preprocessing saliva

```python
all_saliva= []
for i in range(1,23):
    try:
        file_path=BASE_DIR/f"user_{i}"/"saliva.csv"
        df=pd.read_csv(file_path)
        df["user"]=i
        all_saliva.append(df)
    
    except FileNotFoundError:
        print(f"Missing file info for user {i}")
# print(all_saliva)
saliva = pd.concat(all_saliva, ignore_index=True)
print(saliva)
```

```python
saliva2 = (
    saliva
        .set_index("user")
        .drop(columns=["Melatonin NORM"])
        .iloc[:, 1:]
        .rename(columns={"SAMPLES":"time","Cortisol NORM": "cortisol"})
)
saliva2
```
