```python
from EDF_Activity_Analysis import physiological_labelled

agg = (
    physiological_labelled
    .dropna(subset=["activity_label"])
    .groupby(["user", "activity_label"])
    .agg(
        HRV_mean=("HRV", "mean"),
        HRV_median=("HRV", "median"),
        ECG_std=("ECG", "std"),
        acc_x_std=("Accelerometer_X", "std"),
        acc_y_std=("Accelerometer_Y", "std"),
        acc_z_std=("Accelerometer_Z", "std")
    )
    .reset_index()
)
```
